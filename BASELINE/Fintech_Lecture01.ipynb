```python

# curnel density estimation
# 어떤 critical score가 주어졌을 때, 그게 나올 확률을 눈으로 보여주는 함수
# outlier에 영향을 덜 받도록 smoothing을 해줌
# kde = True
fig = plt.figure(figsize=(15, 4))
ax = fig.add_subplot(131)
ax = sns.histplot(data=df, x='Critic_Score', kde=True)
ax2 = fig.add_subplot(132)
ax2 = sns.histplot(data=df, x='Critic_Score', binwidth=1, kde=True)
ax3 = fig.add_subplot(133)
ax3 = sns.histplot(data=df, x='Critic_Score', bins=50, kde=True)
plt.show()
```

```python
# 선그래프로 그리기는 선으로 이어진 두 점 사이에는 선호관계를 암시하는 것을 나타낼 수 있기 때문에 잘못된 이해로 이어질 가능성이 있다.
# categorical data를 선그래프로 그리는 것은 적절하지 않다.
# 선그래프는 연속적인 데이터를 시각화할 때 사용한다.
```
